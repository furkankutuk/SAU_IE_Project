```python
def yorum_temizlenmis(yorum): 
    
    # harflerin ke evrilmesi
    kucult = yorum.str.lower()
    
    # Tekrarlanan karekterlerin deitirilmesi
    karakter_temizle = kucult.str.replace("&#039;", "")
    
    #zel karakterleri kaldrma
    ozel_kar_tem = karakter_temizle.str.replace(r'[^\w\d\s]',' ')
    
    # ASCII karakterleri deitirme
    ascii_temizle = ozel_kar_tem.str.replace(r'[^\x00-\x7F]+',' ')
    
    # bataki ve sondaki Boluklar kaldrma 
    bosluk_kaldir = ascii_temizle.str.replace(r'^\s+|\s+?$','')
    
    # Fazla boluklar tek bolua evirme
    coklu_bosluk = bosluk_kaldir.str.replace(r'\s+',' ')
    
    # 2 ve daha fazla noktalarn tek noktaya dntrlmesi
    dataframe = coklu_bosluk.str.replace(r'\.{2,}', ' ')
    
    return dataframe
```

```python
df3['yorum_temizlenmis'] = yorum_temizlenmis(df3['yorum'])
```

```python
def duygu(yorum):

    polarite = []
    for i in yorum:
        analiz = TextBlob(i)
        polarite.append(analiz.sentiment.polarity)
    return polarite
```

```python
from lightgbm import LGBMModel,LGBMClassifier, plot_importance
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.model_selection import train_test_split
```

```python
ozellikler = df3[['kullanim_alani', 'faydali_bulunma', 'duygu', 'duygu_temizlenmis_ss', 'kelime_sayisi', 
                 'essiz_kelime', 'harf_sayisi', 'noktalama_sayisi', 'buyuk_harfli_kelime', 
                  'etkisiz_kel_sayi', 'ort_kel_uzunlugu']]

hedef = df3['yorum_duygusu']
hedef = hedef.astype('int')

X_train, X_test, y_train, y_test = train_test_split(ozellikler, hedef, test_size = 0.2, random_state = 42)
print ("Eitim veri seti boyutu ", X_train.shape)
print ("Test veri seti boyutu ", X_test.shape)
```

```python
lgbm_sinif = LGBMClassifier(
        n_estimators=10000,
        learning_rate=0.10,
        num_leaves=30,
        subsample=.9,
        max_depth=7,
        reg_alpha=.1,
        reg_lambda=.1,
        min_split_gain=.01,
        min_child_weight=2,
        silent=-1,
        verbose=-1,
        )
egitim = lgbm_sinif.fit(X_train, y_train)

tahminler = egitim.predict(X_test)
print ("Modelin Doruluu : ", accuracy_score(y_test, tahminler), '\n')
print ("Karklk Matrisi \n")
print (confusion_matrix(y_test, tahminler), '\n')

print (classification_report(y_test, tahminler))
```

```python
xgb_sinif = XGBClassifier(n_estimator = 10000,
                    learning_rate=0.10,
                    num_leaves=30)

egitim_xgb = xgb_sinif.fit(X_train, y_train)

tahminler_2 = egitim_xgb.predict(X_test)
print ("Modelin Doruluu : ", accuracy_score(y_test, tahminler_2), '\n')
print ("Karklk Matrisi \n")
print (confusion_matrix(y_test, tahminler_2), '\n')

print (classification_report(y_test, tahminler_2))
```
